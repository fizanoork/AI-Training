```python
#inverse of a metrix

#[[6 1 1],
#[4 -2 5],
#[2 8 7]], 3*3 matrix

import numpy as np
m1 = np.matrix([[6,1,1],[4,-2,5],[2,8,7]])
print(m1)
print(np.linalg.inv(m1))
```

```python
#determinant of matrix

import numpy as np
m1 = np.matrix([[6,1,1],[4,-2,5],[2,8,7]])
print(np.linalg.det(m1))
```

```python
import numpy as np
m1 = np.matrix([[6,1,1],[4,-2,5],[2,8,7]])

from numpy.linalg import det
print(det(m1))
```

```python
#eigen values , eigen vectors
# A = A - lambda I

from numpy.linalg import eig
matrix = np.matrix([[4,2],[3,1]])
values, vectors = eig(matrix)
print(values)
print(vectors)
```

```python
#rank - How rows or columns independent, if it is dependent to any another column,we don't consider the independent.

v1 = np.array([[1,2],[3,4]])
print(np.linalg.matrix_rank(v1))
```
